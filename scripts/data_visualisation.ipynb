Load data, metadata and labels for a specific trial:

```python
trial_dir = "trial42"

data, meta = itemgetter("data", "meta")(rebuild(path2data / trial_dir))

clustering_labels = decompress_binary_matrix(
    meta["labels"]["clustering"], index=data.index, dtype=bool
)
decomposition_labels = decompress_binary_matrix(
    meta["labels"]["decomposition"], index=data.index, dtype=bool
)

display("data", data)
display("clustering_labels", clustering_labels)
display("decomposition_labels", decomposition_labels)
```

Plot the corresponding labels:

```python
ax = plt.matshow(clustering_labels.T, 
            cmap="Reds", 
            aspect="auto", 
            extent=[clustering_labels.index[0], clustering_labels.index[-1], 0, clustering_labels.T.shape[0]],
            origin="lower"
            )
plt.yticks(np.arange(0, clustering_labels.T.shape[0]), clustering_labels.columns)
plt.ylabel("Cluster")
plt.xlabel("Time (s)")
```

```python
ax = plt.matshow(decomposition_labels.T, 
            cmap="Reds", 
            aspect="auto", 
            extent=[decomposition_labels.index[0], decomposition_labels.index[-1], 0, decomposition_labels.T.shape[0]],
            origin="lower"
            )
plt.yticks(np.arange(0, decomposition_labels.T.shape[0]), decomposition_labels.columns)
plt.ylabel("Actuator")
plt.xlabel("Time (s)")
```
